```python
##iterate over squence until condition is met
number =0
while number <5:
    print(number)
    number =number +1
```

```python
##break statement
number =0
while number <5:
    print(number)
    if number ==3:
        break
    number =number +1
```

```python
##break statement
number =0
while number <5:
    print(number)
    if number ==6:
        break
    number =number +1
else:
    print('no longer < 5')
```

```python
##continue statement
number =0
while number <5:
    print(number)
    if number ==3:
        continue
    number =number +1
else:
    print('no longer < 5')
```
